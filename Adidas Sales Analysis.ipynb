```python
#importing python libraries
import pandas as pd
import numpy as np
```

```python
# adidas["Total Sales"] = adidas["Total Sales"].str.replace('.',' ',regex=True)
adidas['Total Sales'] = adidas['Total Sales'].replace(r'[$, ]', '', regex=True).astype(float)
```

```python
adidas.head()
```

```python
#Replacing non numeric characters $
adidas['Units Sold'] = adidas['Units Sold'].replace(r'[$, ]', '', regex=True).astype(float)
adidas['Operating Profit'] = adidas['Operating Profit'].replace(r'[$, ]', '', regex=True).astype(float)
```

```python
adidas.head()
```

```python
adidas.info()
```

```python
#checking datatype
adidas["Invoice Date"].dtypes
```

```python
#Converting Invoice date datatype to_datetime for time series Analysis
adidas["Invoice Date"] = pd.to_datetime(adidas["Invoice Date"])
```

```python
adidas["Invoice Date"].dtypes
```

```python
# Extract Year and Month from Invoice Date
adidas['Year'] = adidas['Invoice Date'].dt.year 
adidas['Month'] = adidas['Invoice Date'].dt.month_name() 
```

```python
adidas.head()
```

```python
#checking for null values in extracted Month & Year column
adidas['Year'].isnull().sum()
adidas['Month'].isnull().sum()
```

```python
#Financial Performance Analysis
#Total Revenue
total_Revenue = adidas['Total Sales'].sum()
total_Revenue
```

```python
#Calculating YOY profit
adidas['YOY_profit']= adidas['Operating Profit'].pct_change(periods=1)*100
adidas['YOY_profit']
```

```python
adidas.head()
```

```python
#.shape gives no of rows & column count
number_of_periods = adidas.shape[0]
number_of_periods
```

```python
#calculating average revenue
Total_revenue = adidas['Total Sales'].sum()
number_of_periods = adidas.shape[0]
Average_revenue = Total_revenue/number_of_periods
Average_revenue
Total_revenue
```

```python
#calculating the revenue for each year
revenue_per_year = adidas.groupby(['Year'])['Total Sales'].sum()
revenue_per_year
```

```python
adidas['YOY_profit_year']= adidas['Year'].pct_change(periods=1)*100
adidas.head()
```

```python
#Calculate revenue for same period last year
revenue_last_year = []
for index, row in adidas.iterrows():
    same_period_last_year = adidas[(adidas['Year'] == row['Year'] - 1) & (adidas['Month'] == row['Month'])]
    revenue_last_year.append(same_period_last_year['Total Sales'].sum())

adidas['Revenue Last Year'] = revenue_last_year
adidas.head()
```

```python
# Calculate average operating margin
average_operating_margin = adidas['Operating Margin'].mean()
average_operating_margin
```
